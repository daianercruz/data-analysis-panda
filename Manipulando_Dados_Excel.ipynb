**Manipulando os Dados com Planilha Excel**

```python
import pandas as pd
```

```python
df1 = pd.read_excel("/content/drive/MyDrive/datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/datasets/Natal.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/datasets/Recife.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/datasets/Salvador.xlsx")
```

```python
df1.head()
```

```python
#juntando os arquivos
df = pd.concat([df1, df2, df3, df4])
```

```python
#Exibindo as 5 primeiras linhas
df.head()
```

```python
#Exibindo as 5 ultimas linhas
df.tail()
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Alterando o tipo de dado da coluna Loja ID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

Tratamento de valores faltantes

```python
#Consultar linhas com valores zerados
df.isnull().sum()
```

```python
#Substituindo os valores zerados pela média
df["Vendas"].fillna(df["Vendas"].mean(),inplace=True)
```

```python
#Deletando linhas que estejam com valores zerados em todas as colunas
df.dropna(how="all", inplace=True)
```

Criando novas colunas

```python
#criando coluna de receita
df["Receita"]= df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#Agrundando por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(15)
```
